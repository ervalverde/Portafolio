# 📊 Análisis de Ventas

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
df.head()
```

```python
ventas_prov = df.groupby('provincia')['monto'].sum().sort_values()

ventas_prov.plot(kind='barh')
plt.title('Ventas por Provincia')
plt.xlabel('Monto')
plt.show()
```

```python
productos = df.groupby('producto')['monto'].sum().sort_values()

productos.plot(kind='bar')
plt.title('Ventas por Producto')
plt.ylabel('Monto')
plt.show()
```

```python
df['fecha'] = pd.to_datetime(df['fecha'])
ventas_fecha = df.groupby('fecha')['monto'].sum()

ventas_fecha.plot()
plt.title('Tendencia de Ventas')
plt.show()
```

```python
ticket_promedio = df['monto'].mean()
print("Ticket promedio:", ticket_promedio)
```

```python
print("Conclusión:")
print("Se observa que ciertas provincias concentran la mayor cantidad de ventas,")
print("y algunos productos dominan el mercado, lo cual puede ayudar a tomar decisiones estratégicas.")
```
